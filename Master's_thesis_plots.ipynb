# **VBA Heatmaps**


pupil heatmap (trigger)

```python
''' to save the workspaces in matlab use save('workspace', '-v6');'''
import scipy.io as sio
import numpy as np
import matplotlib.pyplot as plt
# load mat file

mat_contents = sio.loadmat('workspace_1.mat')
pupil= mat_contents["baseline_pupil"]
vba = mat_contents["baseline"]
cmap =sns.color_palette('inferno',as_cmap=True)
sns.heatmap(pupil,cmap= cmap,alpha = 0.9,vmin = -6,vmax= 6)
#heatmap= ax.pcolor(pupil, cmap=plt.cm.Blues)
#plt.imshow(pupil)
plt.axvline(x = 200, color ="black", linestyle ="--")
#plt.colorbar(shrink=0.5)
plt.title('Heatmap of pupil sizes 10 secs before and after trigger')
plt.xlabel('time')
plt.ylabel('trials')
plt.show()
```

4 conditions  plots

```python
sns.set_theme(style="whitegrid",palette="pastel")
time = np.linspace(1,401,401)
values_trigger= np.mean(pupil,axis =0)
values_replay= np.mean(pupil_replay,axis =0)
values_silent= np.mean(pupil_silent,axis =0)
values_random= np.mean(pupil_random,axis =0)
ci_trigger = 1.96 * np.std(values_trigger)/np.sqrt(len(time))
ci_replay = 1.96 * np.std(values_replay)/np.sqrt(len(time))
ci_silent = 1.96 * np.std(values_silent)/np.sqrt(len(time))
ci_random = 1.96 * np.std(values_random)/np.sqrt(len(time))
plt.plot(time,values_trigger,color ='#005a32')
plt.plot(time,values_replay,color ='#ec7014')
plt.plot(time,values_silent,color ='#feb24c')
plt.plot(time,values_random,color ='#225ea8')
plt.fill_between(time,(values_trigger- ci),(values_trigger +ci),color='#005a32', alpha=0.3)
plt.fill_between(time,(values_replay- ci),(values_replay +ci),color='#ec7014', alpha=0.3)
plt.fill_between(time,(values_silent- ci),(values_silent +ci),color='#feb24c', alpha=0.3)
plt.fill_between(time,(values_random- ci),(values_random +ci),color='#225ea8', alpha=0.3)
plt.axvline(x= 200,linestyle='--',color = 'black')
plt.legend(['trigger','replay','silent','random'])
plt.title('Pupil traces during egress')
plt.xlabel('time')
plt.ylabel('pupil size')
```
